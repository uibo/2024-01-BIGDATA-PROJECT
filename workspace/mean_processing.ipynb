```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager
plt.rc('font', family='NanumGothicCoding')
```

```python
file_name = './data/cleaning_data.csv'
df = pd.read_csv(file_name)
```

```python
mean_price = df['price'].mean()
group = df.groupby('region', as_index=False)['price'].mean()
group.loc[group['region'] == '전국', 'price'] = mean_price
group.rename(columns = {'price':'mean_price'}, inplace = True)
group
```

```python
plt.bar(group['region'], group['mean_price'])
plt.xticks(rotation=45) 

for i, price in enumerate(group['mean_price']):
    plt.text(i, price, str(int(price)), ha='center', va='bottom')

plt.xlabel('Region') 
plt.ylabel('mean_price(원)')  

plt.show()
```

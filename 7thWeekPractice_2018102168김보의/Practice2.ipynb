```python
import pandas as pd
import numpy as np
```

```python
file_name = 'airbnb.csv'
df_air = pd.read_csv(file_name)
df_air['overall_satisfaction'] = df_air.overall_satisfaction.fillna(3)
```

```python
group_mean = df_air.groupby('neighborhood')[['reviews', 'overall_satisfaction', 'price']].mean()
group_mean.columns = ['avg_' + col for col in group_mean.columns]
group_max = df_air.groupby('neighborhood')[['reviews', 'price']].max()
group_max.columns = ['max_' + col for col in group_max.columns]
group_min = df_air.groupby('neighborhood')[['reviews', 'price']].min()
group_min.columns = ['min_' + col for col in group_min.columns]

group = pd.concat([group_mean, group_max, group_min], axis = 1)
group
```

```python
group.to_csv('./sorted_neighborhood_factors.csv')
```

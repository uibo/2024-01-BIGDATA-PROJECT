```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file_name = 'sort_ranged_price.csv'
ad = pd.read_csv(file_name)
file_name = 'sorted_neighborhood_factors.csv'
ad_n = pd.read_csv(file_name)
```

```python
plt.plot(ad['PRICE'], ad['accommodates_average'], marker = 'o', linestyle = '--')
plt.xlabel('ranged_price')
plt.ylabel('accommodates_average')

for i, txt in enumerate(ad['accommodates_average']):
    plt.text(ad['PRICE'][i], ad['accommodates_average'][i], f"{txt:0.3}")
plt.show()
```

```python
plt.plot(ad['PRICE'], ad['bedrooms_average'], marker = 'o', linestyle = '--')
plt.xlabel('ranged_price')
plt.ylabel('bedrooms_average')

for i, txt in enumerate(ad['bedrooms_average']):
    plt.text(ad['PRICE'][i], ad['bedrooms_average'][i], f"{txt:0.3}")
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, sharex =True, figsize=(17, 10))
axs[0].plot(ad_n['neighborhood'], ad_n['avg_reviews'], marker= 'o' , markersize = 4, linestyle = ':')
axs[0].set_ylabel('avg_reviews', fontsize = 14)
axs[1].plot(ad_n['neighborhood'], ad_n['avg_overall_satisfaction'], marker= 'o' , markersize = 4, linestyle = ':')
axs[1].set_ylabel('avg_overall_satisfaction', fontsize = 14)
axs[2].plot(ad_n['neighborhood'], ad_n['avg_price'], marker= 'o' , markersize = 4, linestyle = ':')
axs[2].set_ylabel('avg_price', fontsize = 14)
plt.xlabel('neighborhood', fontsize = 14)
plt.xticks(ad_n['neighborhood'], fontsize=10, rotation=45, ha='right')


plt.show()

```

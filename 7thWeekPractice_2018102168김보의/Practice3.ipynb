```python
import pandas as pd
import numpy as np
```

```python
file_name = 'airbnb.csv'
ad = pd.read_csv(file_name)
ad['overall_satisfaction'] = ad.overall_satisfaction.fillna(3)
```

```python
ad['district'] = 0
ad.loc[ad['price'] <= 100, 'district'] = 0
ad.loc[(ad['price'] > 100) & (ad['price'] <= 200), 'district'] = 1
ad.loc[(ad['price'] > 200) & (ad['price'] <= 300), 'district'] = 2
ad.loc[(ad['price'] > 300) & (ad['price'] <= 400), 'district'] = 3
ad.loc[(ad['price'] > 400) & (ad['price'] <= 500), 'district'] = 4
ad.loc[(ad['price'] > 500) & (ad['price'] <= 1000), 'district'] = 5

air_pricegroup_mean = ad.groupby('district')[['accommodates', 'bedrooms', 'reviews']].mean()
air_pricegroup_mean.columns = [col + '_average' for col in air_pricegroup_mean.columns]
air_pricegroup_median = ad.groupby('district')[['accommodates', 'bedrooms', 'reviews']].median()
air_pricegroup_median.columns = [col + '_median' for col in air_pricegroup_median.columns]
air_pricegroup = pd.concat([air_pricegroup_mean, air_pricegroup_median], axis = 1)
new_order = ['accommodates_average', 'accommodates_median', 'bedrooms_average', 'bedrooms_median', 'reviews_average', 'reviews_median']
air_pricegroup = air_pricegroup[new_order]
air_pricegroup
```

```python
air_pricegroup2 = pd.DataFrame(index = [0,1,2,3,4,5], data = [], columns = ['neighbor_list'])
for i in range(0,6):
    neighbor_list = set(ad.loc[ad['district'] == i, 'neighborhood'])
    air_pricegroup2.at[i,'neighbor_list'] = neighbor_list
air_pricegroup = pd.concat([air_pricegroup, air_pricegroup2], axis = 1)
air_pricegroup
```

```python
air_pricegroup['length'] = None
for i in range(0,6):
    count_district = ad[ad['district'] == i].shape[0]
    air_pricegroup.iloc[i, 7] = count_district
air_pricegroup
```

```python
new_index = ['0-100', '100-200', '200-300', '300-400', '400-500', '500-1000']
air_pricegroup = air_pricegroup.set_index(pd.Index(new_index))
air_pricegroup = air_pricegroup.rename_axis('PRICE')
air_pricegroup
```

```python
air_pricegroup.to_csv('./sort_ranged_price.csv')
```

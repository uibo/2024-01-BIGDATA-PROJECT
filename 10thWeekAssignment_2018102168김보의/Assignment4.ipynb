```python
from pyspark.sql import SparkSession
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hdfs_path = "hdfs://localhost:9000/user/hadoop/"
file_path = hdfs_path + "sorted_ranged_price.csv"
```

```python
spark = SparkSession.builder.appName("SparkonHDFS").getOrCreate()
```

```python
spark_df = spark.read.csv(path=file_path, header=True, inferSchema=True)
```

```python
pandas_df = spark_df.toPandas()
```

```python
plt.figure(figsize=(10,6))
plt.plot(pandas_df['price_range'], pandas_df['accommodates_average'], marker = 'o', linestyle = '--')
plt.xlabel('ranged_price')
plt.ylabel('accommodates_average')
```

```python
plt.figure(figsize=(10,6))
plt.plot(pandas_df['price_range'], pandas_df['bedrooms_average'], marker = 'o', linestyle = '--')
plt.xlabel('ranged_price')
plt.ylabel('bedrooms_average')
```

```python
hdfs_path2 = "hdfs://localhost:9000/user/hadoop/"
file_path2 = hdfs_path + "sorted_neighborhood_factors.csv"
spark_df2 = spark.read.csv(path=file_path2, header=True, inferSchema=True)
ad_n = spark_df2.toPandas()
```

```python
fig, axs = plt.subplots(3, 1, sharex =True, figsize=(17, 10))
axs[0].plot(ad_n['neighborhood'], ad_n['avg of reviews'], marker= 'o' , markersize = 4, linestyle = ':')
axs[0].set_ylabel('avg_reviews', fontsize = 14)
axs[1].plot(ad_n['neighborhood'], ad_n['avg of overall_satisfaction'], marker= 'o' , markersize = 4, linestyle = ':')
axs[1].set_ylabel('avg_overall_satisfaction', fontsize = 14)
axs[2].plot(ad_n['neighborhood'], ad_n['avg of price'], marker= 'o' , markersize = 4, linestyle = ':')
axs[2].set_ylabel('avg_price', fontsize = 14)
plt.xlabel('neighborhood', fontsize = 14)
plt.xticks(ad_n['neighborhood'], fontsize=10, rotation=45, ha='right')


plt.show()
```
